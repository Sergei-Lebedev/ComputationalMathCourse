# Исследовать сходимость метода Эйлера в зависимости от шага интегрирования

1) Необходимо провести иссследование для уравнения 

$$\frac{dx}{dt}=-x$$

Точным решением уравнения является функция

$$x(t)=x_0e^{-t}$$

```python
import numpy as np
import matplotlib.pyplot as plt
import math
from mpl_toolkits.mplot3d import Axes3D

from scipy.spatial.distance import euclidean
```

Определим функцию для нахождения точного решения уравнения

```python
def exact_solution(t):
    return math.exp(-t)
```

И функцию вычисления производной в точке

```python
def f_der(x):
    return -x
```

И функцию решающую уравнение численно с использованием метода Эйлера. Параметры (f0, x0) начальное значение функции в точке, x_finish - задает интервал решения, h - шаг метода. Функция возвращает численное решение уравнения и ошибку по сравнению с точным решением, посчитанную как абсолютное значение разности решений в соответствующей точке.  

```python
def euler(f0,x0,x_finish,h):
    solution = [f0]
    error=[0]
    n = int((x_finish-x0)/float(h));
    x = np.arange(x0, h*n,h)
    for i in range(1,n):
        y = solution[i-1]+f_der(solution[i-1])*h
        solution.append(y)
        error.append(abs(y-exact_solution(x[i])))
    return x, solution, error
```

Проведем три эксперимента с различными значениями шага интегрирования h

1) h=0.01

```python
[x1,y_num1, err1] = euler(1,0,10,0.01)
```

2) h=0.1

```python
[x2,y_num2, err2] = euler(1,0,10,0.1)
```

3) h=0.005

```python
[x3,y_num3, err3] = euler(1,0,10,0.005)
```

```python
[x1,y_num1, err1] = euler(1,0,10,0.01)    
[x2,y_num2, err2] = euler(1,0,10,0.1)
[x3,y_num3, err3] = euler(1,0,10,0.005)

plt.plot(x1,err1, 'r', label="h=0.01")
plt.plot(x2,err2, 'g', label="h=0.1")
plt.plot(x3,err3, 'm', label="h=0.005")
plt.ylabel("Absolute error")
plt.xlabel("t")
plt.title(r"Numerical solution of equation $\frac{dx}{dt}=-x$", fontsize=20)
plt.legend(loc='upper right')
```

На графике выше приведены значения ошибок, полученных при различных значениях шага интегрирования. В силу характеры функции ошибка при значениях t>8 для всех значений шага достаточно мала. 

2) Необходимо провести иссследование для уравнения 

$$\frac{dx}{dt}=x$$

Точным решением уравнения является функция

$$x(t)=x_0e^{t}$$

Как и в предыдущем примере опеределим функции для нахождения точного решения и нахождения производной в точке

```python
def exact_solution2(t):
    return math.exp(t)

def f_der2(x):
    return x

def euler2(f0,x0,x_finish,h):
    solution = [f0]
    error=[0]
    n = int((x_finish-x0)/float(h));
    x = np.arange(x0, h*n,h)
    for i in range(1,n):
        y = solution[i-1]+f_der2(solution[i-1])*h
        solution.append(y)
        error.append(abs(y-exact_solution2(x[i])))
    return x, solution, error
```

Проведем три эксперимента с различными значениями шага интегрирования h (0.1, 0.01, 0.005)

```python
[x1,y_num1, err1] = euler2(1,0,2,0.01)    
[x2,y_num2, err2] = euler2(1,0,2,0.1)
[x3,y_num3, err3] = euler2(1,0,2,0.005)

plt.plot(x1,err1, 'r', label="h=0.01")
plt.plot(x2,err2, 'g', label="h=0.1")
plt.plot(x3,err3, 'm', label="h=0.005")
plt.ylabel("Absolute error")
plt.xlabel("t")
plt.title(r"Numerical solution of equation $\frac{dx}{dt}=x$", fontsize=20)
plt.legend(loc='upper left')
```

На графике выше привидены значения ошибок, полученных при различных значениях шага интегрирования. Так как решением уравнения является сильно возрастающая функция, то начиная с некоторого значения ошибка начинает накпливаться и на графике виден рост с течением времени.

3) Необходимо провести иссследование системы уравений 

$$ \left \lbrace
\begin{array}{ccc} 
\frac{dx}{dt}&=& y \\ 
\frac{dy}{dt}&=& -x 
\end{array} \right .$$

Определим функцию, которая возвращает точное решение

```python
def exact_solution3(t):
    return 5*math.cos(t), -5*math.sin(t)
```

Производная функции в заданной точке

```python
def f_der3(f):
    return np.array([f[1],-f[0]]) 
```

И численную схему метода Эйлера

```python
def euler3(f0,t0,t_finish,h):
    f = [f0]
    n = int((t_finish-t0)/float(h));
    t = np.arange(t0, h*n,h)
    error = [0]
    for i in range(1,n):
        f_t = f[i-1]+f_der3(f[i-1])*h
        f.append(f_t)
        [x_e,y_e] = exact_solution3(t[i])
        error.append(math.sqrt((f_t[0]-x_e)**2+(f_t[1]-y_e)**2))
    return f, t, error
```

Ниже приведен график ошибки, посчитанной как расстояние между точным решение и численным для каждой точки фазового пространства. Значение шага h=0.001

```python
f,t,error = euler3(np.array([5,0]), 0, 10, 0.001)
x = [float(i[0]) for i in f]
y = [float(i[1]) for i in f]

plt.plot(t,error, label="h=0.001")
plt.ylabel("Absolute error")
plt.xlabel("t")
plt.title(r"Error", fontsize=20)
plt.legend(loc='upper left')
```

График хоршо согласуется с теоретическими результатами оценки ошибки метода. Известно, что метод Эйлера является методом первого порядка, и соответственно ошибка интегрирования на интерваля является функцией O(h).

# Решение системы уравнений Рёсслера с испольованием метода Рунге-Кутта 4 порядка.

Система уравнений Рёсслера записывается в виде

$$ \left \lbrace
\begin{array}{ccc} 
\frac{dx}{dt}&=& -y&-&z \\ 
\frac{dy}{dt}&=& x&+&ay\\ 
\frac{dz}{dt}&=& b&+&z(x-c)
\end{array} \right .$$

Запишем метод Рунге-Кутты. Параметры функции: f - правая часть системы уравнений, `(x0, t0)` - начальные условия, h - шаг интегрирования, t1 - интервал

```python
def rk4(f, x0, t0, h, t1):
    n  = x0.size
    m  = int((t1-t0)/float(h))
    vt = [np.zeros(n)]*(m + 1)
    vx = [0]*(m + 1)
    vt[0] = t = t0
    vx[0] = x = x0
    for i in range(1, m + 1):
        k1 = h*f(t, x)
        k2 = h*f(t + 0.5*h, x + 0.5*k1)
        k3 = h*f(t + 0.5*h, x + 0.5*k2)
        k4 = h*f(t + h, x + k3)
        vt[i] = t = t0 + i*h
        vx[i] = x = x + (k1 + k2 + k2 + k3 + k3 + k4)/6
    return vt, vx
```

Правая часть системы уравнений Рёсслера

```python
def rossler(t, vec):
    x,y,z = vec
    return np.array([-y-z, x+a*y, b+z*(x-c)])
```

Зададим параметры `(a,b,c)`. При таких параметрах в систем должен существовать предельный цикл.

```python
a=b=0.1
c=6
```

И начальные условия

```python
initial=[-1,0,0]
```

```python
t,f = rk4(rossler, np.array([-1,0,0]), 0, 0.01, 100)
x = [float(i[0]) for i in f]
y = [float(i[1]) for i in f]
z = [float(i[2]) for i in f]
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot(x, y, z)
```

На графике ниже показана разница между траекториями полученными при различных значениях шага `h`.
Первоначальное значение `h=0.01`, затем шаг уменьшался в 2 раза.

```python
def calcError(x1, x1_2):
    error = []
    for i in range(len(x1)):
        error.append(euclidean(x1[i], x1_2[2*i]))
    return error

t1,f1 = rk4(rossler, np.array([-1,0,0]), 0, 0.01, 100)
t2,f2 = rk4(rossler, np.array([-1,0,0]), 0, 0.005, 100)
t3,f3 = rk4(rossler, np.array([-1,0,0]), 0, 0.0025, 100)

error1 = calcError(f1,f2)
error2 = calcError(f2,f3)
plt.plot(t1,error1, label="$||x_{h}-x_{h/2}||$")
plt.plot(t2,error2, "r", label = "$||x_{h/2}-x_{h/4}||$")
plt.legend(loc='upper left')
plt.title(r"Trajectories difference for different h", fontsize=20)
plt.grid(True)
plt.show()
```
