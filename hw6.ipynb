# Логистическое отображение

Необходимо построить бифуркационную диаграмму и зависимость показателей Ляпунова от параметра для логистического отображения
$$x_{n+1} = rx_n(1-x_n)$$

Определим параметры необходимые для расчета

```python
R = np.linspace(2.1, 4.0, num=3000) #r parameter space
x = 0.1 #Initial value
TR = 500 #Transient response
NumPoints = 1000 #Number of points to draw
```

И функции для вычисления логистического отображения и его производной

```python
def logisticMap(x,r):
    return r * x * (1 -x)
def logisticMapDer(x,r):
    return r-2*r*x
```

Построим бифуркационную диаграмму

```python
xa = []
ra = []
rlam = []
lam = []
for r in R:
    l = 0
    for i in range(TR):
        x = logisticMap(x, r)
        l = l + math.log(abs(logisticMapDer(x, r)))
    for i in range(NumPoints):
        x = logisticMap(x, r)
        l = l + math.log(abs(logisticMapDer(x, r)))
        xa.append(x)
        ra.append(r)
    rlam.append(r)
    lam.append(l / (TR+NumPoints))
    
matplotlib.rcParams.update({'font.size': 22})
pylab.rcParams['figure.figsize'] = (25, 15)
plt.plot(ra,xa, 'o', markersize=0.05)
plt.xlabel("r",fontsize=20)
plt.ylabel("x",fontsize=20)
plt.grid(True)
plt.title(r"Bifurcation diagram", fontsize=30)
```

и зависимость показателей Ляпунова от параметра

```python
matplotlib.rcParams.update({'font.size': 22})
pylab.rcParams['figure.figsize'] = (25, 15)
plt.plot(rlam,lam)
plt.xlabel("r",fontsize=20)
plt.ylabel("$\lambda$",fontsize=20)
plt.grid(True)
plt.title(r"Lyapunov $\lambda$", fontsize=30)
```
