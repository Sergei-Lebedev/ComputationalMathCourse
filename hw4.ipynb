# Исследование закрытых квантовых систем

Зададим матрицу `H`

```python
import numpy as np
def createMatrix(dim):
    H_real = np.random.normal(loc=0.0, scale=1.0, size=(dim,dim))
    H_imag = np.random.normal(loc=0.0, scale=1.0, size=(dim,dim))
    H = H_real + H_imag* 1j
    H = H + H.conjugate().transpose()
    return H
```

```python
H = createMatrix(10)
```

Для матрицы H найдем спектр S

```python
S = np.linalg.eig(H)[0].real
```

А также разницу S_diff между соседними элементами в вариационном ряде составленном по собственным значениям

```python
S_diff = np.diff(np.sort(S))
```

S и S_diff являются реализациями случайной величины. Возьмем несколько реализаций

```python
def findStatistics(N, M):
    final_S = np.array([])
    final_S_diff = np.array([])
    for i in range(M):
        H = createMatrix(N)
        S = np.linalg.eig(H)[0].real
        S_diff = np.diff(np.sort(S))
        final_S = np.append(final_S,S)
        final_S_diff = np.append(final_S_diff,S_diff)
    return final_S, final_S_diff
```
