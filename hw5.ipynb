# Исследование закрытых квантовых систем (ч.2)

1) Для Гауссовых матриц вида
$$H(t) = H_0+H_1cos(\omega t)$$ $$где H_0, H_1 - гауссовы, w=\frac{2\pi}{T}, T = 2\pi$$

Реализовать процедуру отыскания собственных чисел и собственных векторов Флоке

Зададим функцию генерации гауссовых матриц:

```python
import numpy as np
def createMatrix(dim):
    H_real = np.random.normal(loc=0.0, scale=1.0, size=(dim,dim))
    H_imag = np.random.normal(loc=0.0, scale=1.0, size=(dim,dim))
    H = H_real + H_imag* 1j
    H = H + H.conjugate().transpose()
    return H
```

И функцию вычисляющуя оператор Гамильтона H

```python
def H(H0, H1, t):
    return H0+H1*np.cos(t)
```

Исходя из N начальных условий вида $$\psi^k_n=\delta_{nk}$$ найдем численное решение уравнения Шредингера $$i\dot{\psi}=H(t)\psi$$

Для этого зададим функцию генерирующую начальное условие указанного вида

```python
def getY0(dim, k):
    a = np.zeros(dim)
    a[k] = 1
    return a
```

А также функцию для нахождения численного решения для данного оператора H

```python
from math import pi
from scipy.integrate import ode

def getSolution(H0, H1, T, y0, dt):
    def f(t, y, args):
        H0 = args[0]
        H1 = args[1]
        return (-1j*(H0+H1*np.cos(t)).dot(y))
    r = ode(f).set_integrator('zvode', with_jacobian=False)
    args = [H0, H1]
    t0 = 0
    r.set_initial_value(y0, t0).set_f_params(args)
    while r.successful() and r.t < T:
        r.integrate(r.t+dt)
    return r.y
```

Зададим функцию для нахождения оператора эволюции. В качестве аргументов функция будет принимать размерность системы и шаг интегрирования

```python
def constructEvolutionOperator(N, dt):
    H0 = createMatrix(N)
    H1 = createMatrix(N)
    U = []
    T = 2 * pi
    for i in range(N):
        y0 = getY0(N, i)
        y = getSolution(H0, H1, T, y0, dt)
        U.append(y)
    return np.matrix(U).transpose(), H0, H1
```

Определим функции для оценки погрешности

$$E_{abs}=max{|1-|\mu_n||}$$

```python
def E_ABS(eigvals):
    error = [abs(1-abs(values)) for values in eigvals]
    return max(error)
```

$$e_{abs} = |<\phi^k, \frac{1}{\mu_k}\overline{\phi^{k*}}>|$$
$$e_{arg} = arg(<\phi^k, \frac{1}{\mu_k}\overline{\phi^{k*}}>)$$

```python
def e_abs_arg(H0, H1, eigvals, eigvecs, dt, T):
    r_abs = []
    r_arg = []
    for i in range(len(eigvals)):
        v = np.squeeze(np.asarray(eigvecs[:,i]))
        y = getSolution(H0, H1, T, v, dt)
        E = np.dot(v, (y/eigvals[i]).conj().transpose())
        r_abs.append(abs(abs(E)-1))
        r_arg.append(abs(np.angle(E)))
    return np.array(r_abs), np.array(r_arg)
```

Выполним вычисления для N=100, h=0.01

```python
N = 100
dt = 0.01
T = 2*pi
U,H0,H1= constructEvolutionOperator(N, dt)
```

Найдем собственные значения и вектора оператора эволюции

```python
eigvals, eigvecs =  np.linalg.eig(U)
```

```python
Error_abs = E_ABS(eigvals)
```

```python
er_abs, er_arg = e_abs_arg(H0,H1, eigvals, eigvecs, dt, T)
```
